# Load necessary libraries

```python
# Import pandas library for data reading and manipulation
import pandas as pd

# Import numpy for maths functions 
import numpy as np

# Import to visualise decision tree
import pydotplus
import graphviz

# Import decision tree function from sklearn
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

# Read the HR data into Jupyter notebook

```python
# Use the read_csv function in Pandas to read the HR file into Python
hr = pd.read_csv('HR_data - HR_data.csv')
```

```python
# Let's see what our data looks like
hr
```

# Build decision tree model!

```python
dtree = DecisionTreeClassifier(max_depth= 3)
dtree.fit(hr.drop(['resigned'], axis= 1), hr['resigned'])
```

```python
# Let's define a function for plotting our decision tree!
dot_data = export_graphviz(dtree, 
                                out_file=None, 
                                feature_names=hr.drop(['resigned'], axis =1).columns,
                                filled=True, 
                                rounded=True)
graph = graphviz.Source(dot_data)
```
